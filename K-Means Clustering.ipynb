## From given 'Iris' dataset,predicting optimum number of clusters and representing it visually

### Loading Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn as sk
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

## Define your target and predictions

```python
df=pd.DataFrame(data=iris.data,columns=['sepal length','sepal width','petal length','petal width'])
df['target']=pd.Series(iris.target)
df['target_names']=pd.Series(iris.target_names)
species=[]
for i in range(len(df)):
    if df.iloc[i]['target']==0:
        species.append('setosa')
    elif df.iloc[i]['target']==1:
        species.append('versicolor')
    elif df.iloc[i]['target']==2:
        species.append('virginica')
df['Species']=species
```

```python
df
```

```python
x=iris.data
x
```

```python
plt.scatter(x=df['sepal length'],y=df["sepal width"],c=iris.target,cmap="gist_rainbow")
plt.xlabel('Sepal length')
plt.ylabel('sepal width')
```

## Using elbow method to estimate the value k

```python
err=[]
for i in range(1,11):
    kmeans11=KMeans(n_clusters = i, init = 'k-means++', max_iter=300, n_init=10, random_state=0).fit(x)
    kmeans11.fit(x)
    err.append(kmeans11.inertia_)

plt.plot(range(1,11),err)
plt.title('Elbow-Method using  wcss with k=1-11')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

#### As you can see the optimal value of k is between 2 and 4, as elbow-like shape is formed at k=3 in the above graph
#### Lets Implement k=means using k=3


```python
kmeans3=KMeans(n_clusters=3, random_state=42)
y=kmeans3.fit_predict(x)
print(y)
```

#### Indentify the center points of clusters


```python
kmeans3.cluster_centers_
```

```python
kmeans= KMeans(n_clusters=3, init ="k-means++",max_iter=100,n_init=10, random_state=0)
y_kmeans= kmeans.fit_predict(x)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x[y_kmeans ==0,0],x[y_kmeans==0,1],s=70,c='red',label="iris_setosa")
plt.scatter(x[y_kmeans ==1,0],x[y_kmeans==1,1],s=70,c='blue',label="iris_versicolour")
plt.scatter(x[y_kmeans ==2,0],x[y_kmeans==2,1],s=70,c='green',label="iris_virginica")

#Centroid of Clusters
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=70,c="black",label="Centroids")
plt.legend()
plt.plot(figsize=(60,40))
```
